```python
import spacy
import pandas as pd
nlp = spacy.load('en_core_web_sm')
```

```python
string = "I am a runner running in a race because I love to run since I ran today."
doc = nlp(string)
```

```python
pd.DataFrame(
    {
        'token': [token.text for token in doc],
        'pos': [token.pos_ for token in doc],
        'lemma': [token.lemma_ for token in doc],
        'lemma_code': [token.lemma for token in doc],
    }
)
    
```
